```python
from sksurv.ensemble import RandomSurvivalForest
from sksurv.linear_model import CoxPHSurvivalAnalysis
```

```python
SELECTED_sur_FEATURES = ["age", "anemic", "aspartate aminotransferase", "cardiopathic", "chlorine",
                        "creatinine", "creatinine delta","diabetic", "erythrocytes", "erythrocytes delta", 
                        "GFR delta last 4 months", "GFR delta last year", "GFR standard deviation last 4 months",
                        "GFR standard deviation last year", "hematocrit", "hemoglobin", "hypertension episodes",
                        "last GFR", "male", "mean corpuscular hemoglobin", "mean corpuscular volume", "potassium",
                        "sodium", "specific gravity standard deviation", "urate", "urea", "urea delta",'months until dialysis','EVENT']
```

```python
all_data_surv = all_data[SELECTED_sur_FEATURES[:-2]]
```

```python
all_data_surv
```

```python
all_data['EVENT']=True
```

```python
y =all_data[['months until dialysis','EVENT']]
```

```python
y
```

```python
rsf = RandomSurvivalForest(n_estimators=1000,
                           min_samples_split=10,
                           min_samples_leaf=15,
                           max_features="sqrt",
                           n_jobs=-1,
                           random_state=39)
```

```python
from  sksurv import util
```

```python
y = util.Surv.from_dataframe('EVENT','months until dialysis',y) 
```

```python
y
```

```python
X_train,X_test, y_train, y_test = train_test_split(all_data_surv,y,random_state=39)
```

```python
rsf.fit(X_train,y_train)
```

```python
rsf.score(X_test,y_test)
```

```python
surv = rsf.predict_survival_function(X_test.head(), return_array=True)

for i, s in enumerate(surv):
    plt.step(rsf.event_times_, s, where="post", label=str(i))
plt.ylabel("Dialysis probability")
plt.xlabel("Time in month")
plt.legend()
plt.grid(True)
```

```python
surv = rsf.predict_cumulative_hazard_function(X_train.head(), return_array=True)

for i, s in enumerate(surv):
    plt.step(rsf.event_times_, s, where="post", label=str(i))
plt.ylabel("Cumulative hazard")
plt.xlabel("Time in month")
plt.legend()
plt.grid(True)
```

```python
import eli5
from eli5.sklearn import PermutationImportance

perm = PermutationImportance(rsf, n_iter=15, random_state=39)
perm.fit(X_test, y_test)
eli5.show_weights(perm, feature_names=X.columns.to_list())
```
